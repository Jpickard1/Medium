```python
# Numpy is the only library needed for the perceptron algorithm
import numpy as np
# Imports for generating data
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
# Used for plotting
import matplotlib.pyplot as plt
```

## Perceptron Algorithm

```python
def train_perceptron(X, y):
  theta = np.zeros(len(X[0]))
  misclassified = True
  while misclassified:
    misclassified = False
    for i in range(len(X)):
      if y[i] * np.dot(theta, X[i]) <= 0:
        theta += y[i] * X[i]
        misclassified = True
  return theta
```

```python
def predict(theta, x):
  return np.sign(np.dot(theta, x))
```

### Generate Simulated Data

```python
# Create data
X, y = make_blobs(n_samples=1000, n_features=2, centers=2, shuffle=True, random_state=13)
# Set labels to 1, -1 by replacing 0s with -1s
y[np.where(y==0)] = -1
# Plot the data
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Feature Space')
plt.show()
```

### Train the Perceptron Algorithm

```python
theta = train_perceptron(X,y)
```

### Plot the Decision Boundary

```python
# Calculate the decision boundary
x1 = [0, 0]
x2 = [1, theta[0]/theta[1]]
slope = theta[0]/theta[1]

pt1 = [min(X[:,0]), max(X[:,0])]
pt2 = [pt1[0]*slope, pt1[1]*slope]

# Plot data
plt.scatter(X[:, 0], X[:, 1], c=y)
# Plot decision boundary
plt.plot(pt1, pt2, 'y-')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Feature Space')
plt.show()
```

## Improvements to the Perceptron

```python
def train_perceptron(X, y, max_epochs=10):
  theta = np.zeros(len(X[0]))
  b = 0
  misclassified = True
  epoch = 0
  while misclassified and epoch < max_epochs:
    epoch += 1
    misclassified = False
    for i in range(len(X)):
      if y[i] * (np.dot(theta, X[i])) + b <= 0:
        theta += y[i] * X[i]
        b += y[i]
        misclassified = True
  return theta, b
```

```python
# Create data
X, y = make_blobs(n_samples=1000, n_features=2, centers=2, shuffle=True, random_state=2)
# Set labels to 1, -1 by replacing 0s with -1s
y[np.where(y==0)] = -1
# Plot the data
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Feature Space')
plt.show()
```

```python
theta, b = train_perceptron(X,y)
```

```python
# Calculate the decision boundary
slope = -theta[1]/theta[0]
pt1 = [min(X[:,0]), max(X[:,0])]
pt2 = [pt1[0]*slope, pt1[1]*slope]
pt2 += (-b/theta[0])

# Plot data
plt.scatter(X[:, 0], X[:, 1], c=y)
# Plot decision boundary
plt.plot(pt1, pt2, 'y-')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Feature Space')
plt.show()
```
